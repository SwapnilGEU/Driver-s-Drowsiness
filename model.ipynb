```python
import time
import cv2
from playsound import playsound
from ultralytics import YOLO  
import threading

model = YOLO("y_new2.pt")

cap = cv2.VideoCapture(0)

drowsy_detected_time = None
alert_played = False
alert_threshold = 3  

frame_count = 0
fps = 0
start_time = time.time()

def play_alert():
    playsound("mixkit-rooster-crowing-in-the-morning-2462.wav")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    results = model(frame)[0]
    class_names = [model.names[int(cls)] for cls in results.boxes.cls]

    elapsed_time = time.time() - start_time
    if elapsed_time > 1:
        fps = frame_count / elapsed_time
        frame_count = 0
        start_time = time.time()

    if class_names.count('Closed Eye') >= 2:
        if drowsy_detected_time is None:
            drowsy_detected_time = time.time()
        else:
            elapsed = time.time() - drowsy_detected_time
            if elapsed >= alert_threshold and not alert_played:
                threading.Thread(target=play_alert).start()
                alert_played = True
    else:
        drowsy_detected_time = None
        alert_played = False

    annotated_frame = results.plot()

    cv2.putText(annotated_frame, f"FPS: {fps:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow("Drowsiness Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
